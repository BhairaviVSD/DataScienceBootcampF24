1. Filter the data to include only weekdays (Monday to Friday) and plot a line graph showing the pedestrian counts for each day of the week.

```python
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
df_weekdays = df[df['day_name'].isin(weekdays)]
df_weekdays
```

```python
daily_counts = df_weekdays.groupby('day_name')['Pedestrians'].mean().reindex(weekdays)
daily_counts
```

```python
plt.figure(figsize=(10, 6))
plt.plot(daily_counts.index, daily_counts.values, marker='o')
plt.title('Average Pedestrian Counts by Weekday')
plt.xlabel('Day of the Week')
plt.ylabel('Average Pedestrian Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

2. Track pedestrian counts on the Brooklyn Bridge for the year 2019 and analyze how different weather conditions influence pedestrian activity in that year. Sort the pedestrian count data by weather summary to identify any correlations( with a correlation matrix) between weather patterns and pedestrian counts for the selected year.

- This question requires you to show the relationship between a numerical feature(Pedestrians) and a non-numerical feature(Weather Summary). In such instances we use Encoding. Each weather condition can be encoded as numbers( 0,1,2..). This technique is called One-hot encoding.

- Correlation matrices may not always be the most suitable visualization method for relationships involving categorical datapoints, nonetheless this was given as a question to help you understand the concept better.

```python
df_2019 = df[(df['hour_beginning'].dt.year == 2019) & (df['location'] == 'Brooklyn Bridge')]
df_2019
```

```python
weather_encoded = pd.get_dummies(df_2019['weather_summary'], prefix='weather')
df_encoded = pd.concat([df_2019, weather_encoded], axis=1)
```

```python
weather_encoded
```

```python
df_encoded
```

```python
weather_avg = df_encoded.groupby('weather_summary')['Pedestrians'].mean().sort_values(ascending=False)
weather_avg
```

```python
corr_matrix = df_encoded[['Pedestrians'] + list(weather_encoded.columns)].corr()
corr_matrix
```

```python
plt.figure(figsize=(12, 6))
weather_avg.plot(kind='bar')
plt.title('Average Pedestrian Count by Weather Condition (Brooklyn Bridge, 2019)')
plt.xlabel('Weather Condition')
plt.ylabel('Average Pedestrian Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix: Pedestrians vs Weather Conditions (Brooklyn Bridge, 2019)')
plt.tight_layout()
plt.show()
```

```python
print("Top 5 weather conditions with highest average pedestrian counts:")
print(weather_avg.head())
```

```python
print("Correlation between pedestrian counts and weather conditions:")
print(corr_matrix['Pedestrians'].sort_values(ascending=False))
```

3. Implement a custom function to categorize time of day into morning, afternoon, evening, and night, and create a new column in the DataFrame to store these categories. Use this new column to analyze pedestrian activity patterns throughout the day.

```python
def categorize_time(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['time_category'] = df['hour_beginning'].dt.hour.apply(categorize_time)

time_category_avg = df.groupby('time_category')['Pedestrians'].mean().reindex(['Morning', 'Afternoon', 'Evening', 'Night'])
time_category_avg
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=time_category_avg.index, y=time_category_avg.values)
plt.title('Average Pedestrian Count by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Average Pedestrian Count')
plt.show()
```

```python
hourly_avg = df.groupby([df['hour_beginning'].dt.hour, 'time_category'])['Pedestrians'].mean().unstack()
hourly_avg
```

```python
plt.figure(figsize=(12, 6))
for category in ['Morning', 'Afternoon', 'Evening', 'Night']:
    plt.plot(hourly_avg.index, hourly_avg[category], marker='o', label=category)

plt.title('Average Pedestrian Count by Hour and Time Category')
plt.xlabel('Hour of Day')
plt.ylabel('Average Pedestrian Count')
plt.xticks(range(0, 24))
plt.legend(title='Time Category')
plt.grid(True)
plt.tight_layout()
plt.show()
```
