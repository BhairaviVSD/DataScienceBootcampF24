1. Filter the data to include only weekdays (Monday to Friday) and plot a line graph showing the pedestrian counts for each day of the week.

```python
# Plotting a line plot of pedestrian counts over time
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
filtered_df = df[df['day_name'].isin(weekdays)]

plt.figure(figsize=(12, 6))
plt.plot(filtered_df['day_name'], filtered_df['Pedestrians'], color='blue')
plt.title('Pedestrian Counts Over Weekdays')
plt.xlabel('Weekday')
plt.ylabel('Pedestrian Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting a line plot of pedestrian counts over time, ordered by weekdays
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
filtered_df = df[df['day_name'].isin(weekdays)]

# Ensure the correct ordering of the weekdays
filtered_df['day_name'] = pd.Categorical(filtered_df['day_name'], categories=weekdays, ordered=True)
filtered_df = filtered_df.sort_values('day_name')

plt.figure(figsize=(12, 6))
plt.plot(filtered_df['day_name'], filtered_df['Pedestrians'], color='blue')
plt.title('Pedestrian Counts Over Weekdays')
plt.xlabel('Weekday')
plt.ylabel('Pedestrian Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
filtered_df = df[df['day_name'].isin(weekdays)]

# Group by 'day_name' and calculate the mean pedestrian count for each day
grouped_df = filtered_df.groupby('day_name')['Pedestrians'].mean().reindex(weekdays)

# Plotting the line plot of pedestrian counts for each weekday
plt.figure(figsize=(10, 6))
plt.plot(grouped_df.index, grouped_df.values, marker='o', color='blue')

# Adding title and labels
plt.title('Pedestrian Counts Over Weekdays')
plt.xlabel('Weekday')
plt.ylabel('Average Pedestrian Count')
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
weekdays = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
filtered_df = df[df['day_name'].isin(weekdays)]

# Group by 'day_name' and calculate the total pedestrian count for each day
grouped_df = filtered_df.groupby('day_name')['Pedestrians'].sum().reindex(weekdays)

# Plotting the line plot of pedestrian counts for each weekday
plt.figure(figsize=(10, 6))
plt.plot(grouped_df.index, grouped_df.values, marker='o', color='blue')

# Adding title and labels
plt.title('Total Pedestrian Counts Over Weekdays')
plt.xlabel('Weekday')
plt.ylabel('Total Pedestrian Count')
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```

2. Track pedestrian counts on the Brooklyn Bridge for the year 2019 and analyze how different weather conditions influence pedestrian activity in that year. Sort the pedestrian count data by weather summary to identify any correlations( with a correlation matrix) between weather patterns and pedestrian counts for the selected year.

- This question requires you to show the relationship between a numerical feature(Pedestrians) and a non-numerical feature(Weather Summary). In such instances we use Encoding. Each weather condition can be encoded as numbers( 0,1,2..). This technique is called One-hot encoding.

- Correlation matrices may not always be the most suitable visualization method for relationships involving categorical datapoints, nonetheless this was given as a question to help you understand the concept better.

```python
import seaborn as sns


# 2. Filter data for the year 2019 and Brooklyn Bridge location
df_2019 = df[(df['hour_beginning'].dt.year == 2019) & (df['location'] == 'Brooklyn Bridge')]

# 3. Drop rows with missing values in 'weather_summary' or any relevant columns
df_2019 = df_2019.dropna(subset=['weather_summary', 'Pedestrians'])

# 4. Perform One-hot Encoding for 'weather_summary'
encoded_weather = pd.get_dummies(df_2019['weather_summary'], prefix='Weather')

# 5. Combine pedestrian counts and the encoded weather data
df_2019_encoded = pd.concat([df_2019[['Pedestrians']], encoded_weather], axis=1)

# 6. Generate the correlation matrix
corr_matrix = df_2019_encoded.corr()

# 7. Plot the correlation matrix using Seaborn heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix: Pedestrian Counts and Weather Conditions (2019)')
plt.show()
```
